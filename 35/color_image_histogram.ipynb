```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image , ImageDraw, ImageFont ,ImageOps
from bidi.algorithm import get_display
import arabic_reshaper
```

```python
#Read a color image with PIL.
img = Image.open('Input/1.jpg')
plt.imshow(img)
plt.show()
```

```python
# Write a متن فارسی on image.
reshaped_text = arabic_reshaper.reshape('سلام دنیا')    
bidi_text = get_display(reshaped_text)           

font = ImageFont.truetype('Input/Dana-Medium.ttf', size=40)
draw = ImageDraw.Draw(img)
draw.text((250, 20), bidi_text, (255,255,255), font=font)
draw = ImageDraw.Draw(img)

plt.imshow(img)
plt.show()
```

```python
#Calculate 3 histograms and show with plt.
colors = ['red' , 'green' , 'blue']
for i,color in enumerate(colors):
    img_array = np.array(img)
    lum_img = img_array[:,:,i]
    plt.hist(lum_img.ravel(), bins=range(256),color = color)
plt.show()
```

```python
#Equalizes the image histogram.
img_eq = ImageOps.equalize(img) 
plt.imshow(img_eq)
plt.show()

for i,color in enumerate(colors):
    img_array_eq = np.array(img_eq)
    lum_img_eq = img_array_eq[:,:,i]
    plt.hist(lum_img_eq.ravel(), bins=range(256),color = color)
plt.show()
```

```python
#Convert image to gray.
result = np.mean(img,axis=2)
plt.imshow(result,cmap='gray')
plt.show()
```

```python
#Calculate histogram and show with plt.
plt.hist(result.ravel(), bins=range(256),color = color)
plt.show()
```
