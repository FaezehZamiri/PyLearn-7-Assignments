```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('input/ballons.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
H,S,V = cv2.split(img)
```

```python
lower_hue = np.array([165,0,0]) 
upper_hue = np.array([180,255,255])
mask = cv2.inRange(img, lower_hue, upper_hue)
```

```python
plt.subplot(1,2,1)
img_rgb = cv2.cvtColor(img,cv2.COLOR_HSV2RGB)
plt.imshow(img_rgb)

plt.subplot(1,2,2)
plt.imshow(mask,cmap='gray')

plt.show()
```
