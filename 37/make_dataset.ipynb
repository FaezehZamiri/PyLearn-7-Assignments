```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('Input/numbers.webp')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
rows,cols,bands = img.shape
k = 0
count = 0
for i in range(0,rows,20):
    for j in range(0,cols,20):
        count += 1
        num = img[i:i+20,j:j+20]
        filepath = f"Output/{k}/{count}.jpg"
        cv2.imwrite(filepath,num)
        if np.mod(count,500) == 0 :
            k += 1
```
